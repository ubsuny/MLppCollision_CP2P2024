## Area of circle using Monte Carlo

```python
import random
import matplotlib.pyplot as plt

def monte_carlo_circle(radius, num_points):
    """
    Estimate the area of a circle using Monte Carlo simulation.

    Parameters:
    - radius (float): The radius of the circle.
    - num_points (int): The number of points to use in the simulation.

    Returns:
    - area_circle (float): Estimated area of the circle.
    - circle_points_x (list): X-coordinates of points inside the circle.
    - circle_points_y (list): Y-coordinates of points inside the circle.
    - square_points_x (list): X-coordinates of points outside the circle.
    - square_points_y (list): Y-coordinates of points outside the circle.
    """
    inside_circle = 0
    circle_points_x = []
    circle_points_y = []
    square_points_x = []
    square_points_y = []
    
    # Perform Monte Carlo simulation
    for _ in range(num_points):
        x = random.uniform(-radius, radius)
        y = random.uniform(-radius, radius)
        
        # Check if the point falls inside the circle
        if x**2 + y**2 <= radius**2:
            inside_circle += 1
            circle_points_x.append(x)
            circle_points_y.append(y)
        else:
            square_points_x.append(x)
            square_points_y.append(y)
    
    # Calculate area of the square and estimate area of the circle
    area_square = (2 * radius) ** 2
    area_circle = (inside_circle / num_points) * area_square
    
    return area_circle, circle_points_x, circle_points_y, square_points_x, square_points_y

# Input radius and number of points
radius = float(input("Enter the radius of the circle: "))
num_points = int(input("Enter the number of points to use in the simulation: "))

# Run Monte Carlo simulation
estimated_area, circle_points_x, circle_points_y, square_points_x, square_points_y = monte_carlo_circle(radius, num_points)

# Plotting
plt.figure(figsize=(8, 8))
plt.scatter(circle_points_x, circle_points_y, color='blue', s=5, label='Points inside circle')
plt.scatter(square_points_x, square_points_y, color='red', s=5, label='Points outside circle')
plt.gca().set_aspect('equal', adjustable='box')
circle = plt.Circle((0, 0), radius, color='black', fill=False, linestyle='--', label='Circle')
plt.gca().add_patch(circle)
plt.title('Monte Carlo Simulation for Estimating Circle Area')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Print estimated area of the circle
print(f"Estimated area of the circle using Monte Carlo simulation: {estimated_area}")
```
